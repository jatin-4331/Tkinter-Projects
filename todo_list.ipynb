# Tkinter To-Do List
Run the cell below to launch the to-do list GUI.

```python
import tkinter as tk
from tkinter import messagebox

def add_task():
    task = task_entry.get()
    if task != "":
        task_listbox.insert(tk.END, task)
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "You must enter a task.")

def delete_task():
    try:
        selected_task_index = task_listbox.curselection()[0]
        task_listbox.delete(selected_task_index)
    except IndexError:
        messagebox.showwarning("Warning", "You must select a task to delete.")

def clear_tasks():
    task_listbox.delete(0, tk.END)

# Create the main window
root = tk.Tk()
root.title("To-Do List")
root.geometry("350x450")
root.config(bg="#f0f0f0")

# Set up the title label
title_label = tk.Label(root, text="My To-Do List", font=("Helvetica", 18, "bold"), bg="#f0f0f0")
title_label.pack(pady=10)

# Set up the entry field
task_entry = tk.Entry(root, font=("Helvetica", 14), width=24)
task_entry.pack(pady=10)

# Set up the buttons frame
button_frame = tk.Frame(root, bg="#f0f0f0")
button_frame.pack(pady=10)

add_button = tk.Button(button_frame, text="Add Task", font=("Helvetica", 12), width=10, bg="#4caf50", fg="white", command=add_task)
add_button.grid(row=0, column=0, padx=5)

delete_button = tk.Button(button_frame, text="Delete Task", font=("Helvetica", 12), width=10, bg="#f44336", fg="white", command=delete_task)
delete_button.grid(row=0, column=1, padx=5)

# Set up the listbox
task_listbox = tk.Listbox(root, font=("Helvetica", 14), width=25, height=10, selectbackground="#a6a6a6")
task_listbox.pack(pady=10)

# Set up the clear button
clear_button = tk.Button(root, text="Clear All", font=("Helvetica", 12), width=22, bg="#ff9800", fg="white", command=clear_tasks)
clear_button.pack(pady=5)

# Run the application
root.mainloop()
```
